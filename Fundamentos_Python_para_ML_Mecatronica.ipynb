## 1. Variables y tipos de datos

Una **variable** es un espacio donde almacenamos información.  
En ingeniería, una variable puede representar:

- la temperatura medida por un sensor,
- el voltaje de una batería,
- la velocidad de un motor,
- la posición de un actuador.

### Tipos de datos comunes
- `int`: enteros
- `float`: números decimales
- `str`: texto
- `bool`: verdadero o falso

````python
temperatura = 24.7           # float
rpm_motor = 1500            # int
estado_sensor = True        # bool
nombre_equipo = "Brazo robot"  # str

print("Temperatura:", temperatura)
print("RPM del motor:", rpm_motor)
print("Estado del sensor:", estado_sensor)
print("Equipo:", nombre_equipo)

print(type(temperatura))
print(type(rpm_motor))
print(type(estado_sensor))
print(type(nombre_equipo))
````

### Conversión de tipos

A veces un dato llega como texto desde un archivo CSV, puerto serial o interfaz HMI, y necesitamos convertirlo a número.

````python
lectura_serial = "36.5"
temperatura_convertida = float(lectura_serial)

print("Dato original:", lectura_serial, type(lectura_serial))
print("Dato convertido:", temperatura_convertida, type(temperatura_convertida))
````

## 2. Operadores aritméticos y comparaciones

Python permite realizar operaciones matemáticas de forma directa.

### Operadores aritméticos
- `+` suma
- `-` resta
- `*` multiplicación
- `/` división
- `**` potencia
- `%` residuo

````python
voltaje = 12
corriente = 2.5
potencia = voltaje * corriente

print("Potencia eléctrica =", potencia, "W")

distancia = 8
tiempo = 2
velocidad = distancia / tiempo
print("Velocidad =", velocidad, "m/s")

print("2 elevado a la 3 =", 2**3)
print("17 mod 5 =", 17 % 5)
````

### Operadores de comparación

Se usan mucho para control, validación y toma de decisiones.

````python
temperatura_motor = 72

print("¿Temperatura mayor a 70?", temperatura_motor > 70)
print("¿Temperatura igual a 60?", temperatura_motor == 60)
print("¿Temperatura menor o igual a 80?", temperatura_motor <= 80)
````

## 3. Estructuras condicionales: `if`, `elif`, `else`

Las condicionales permiten que el programa tome decisiones.  
Esto es esencial en Mecatrónica para:

- activar alarmas,
- encender ventiladores,
- detener motores,
- clasificar estados del sistema.

````python
temperatura_driver = 78

if temperatura_driver < 50:
    print("Sistema en condición normal")
elif temperatura_driver < 70:
    print("Sistema en condición de atención")
else:
    print("ALERTA: temperatura elevada, revisar disipación")
````

### Ejemplo aplicado: decisión simple en un sensor ultrasónico

````python
distancia_cm = 12

if distancia_cm < 10:
    print("Objeto demasiado cerca")
else:
    print("Distancia segura")
````

## 4. Ciclos: `for` y `while`

Los ciclos son muy útiles para:
- procesar varias lecturas de sensores,
- repetir cálculos,
- recorrer listas de mediciones,
- automatizar análisis de datos.

### Ciclo `for`
Se usa cuando queremos recorrer una secuencia de datos.

````python
corrientes = [1.8, 2.0, 2.1, 1.9, 2.2]

for i, valor in enumerate(corrientes):
    print(f"Lectura {i}: {valor} A")
````

### Ciclo `while`
Se usa cuando queremos repetir una acción mientras se cumpla una condición.

````python
contador = 0

while contador < 5:
    print("Iteración:", contador)
    contador += 1
````

## 5. Funciones

Una **función** permite encapsular instrucciones para reutilizarlas.  
En ingeniería esto ayuda a mantener programas más organizados y fáciles de depurar.

### Sintaxis general
```python
def nombre_funcion(parametros):
    instrucciones
    return resultado
```

````python
def calcular_potencia(voltaje, corriente):
    potencia = voltaje * corriente
    return potencia

p = calcular_potencia(24, 3.2)
print("Potencia calculada:", p, "W")
````

### Ejemplo aplicado: conversión de ADC a voltaje

Supongamos un ADC de 10 bits con referencia de 5 V.

````python
def adc_a_voltaje(valor_adc, vref=5.0, resolucion=1023):
    return (valor_adc / resolucion) * vref

adc = 512
voltaje_sensor = adc_a_voltaje(adc)
print("Voltaje estimado:", round(voltaje_sensor, 3), "V")
````

## 6. Listas, tuplas y diccionarios

Estas estructuras permiten organizar información.

### Listas
Son colecciones ordenadas y modificables.

````python
temperaturas = [22.4, 22.8, 23.1, 23.7, 24.0]
print("Lista completa:", temperaturas)
print("Primer valor:", temperaturas[0])
print("Último valor:", temperaturas[-1])

temperaturas.append(24.3)
print("Lista actualizada:", temperaturas)
````

### Tuplas
Son parecidas a las listas, pero no se modifican fácilmente.

````python
robot_pose = (120.0, 45.0, 300.0)  # x, y, z
print("Posición del robot:", robot_pose)
````

### Diccionarios
Son útiles cuando queremos asociar un nombre con un valor.

````python
sensor = {
    "id": "TMP_01",
    "temperatura_C": 26.5,
    "ubicacion": "Gabinete de control",
    "activo": True
}

print(sensor)
print("Temperatura:", sensor["temperatura_C"], "°C")
````

## 7. Arreglos con NumPy

Para ML, las listas de Python no siempre son suficientes.  
`NumPy` permite trabajar de manera eficiente con arreglos numéricos.

Esto es clave cuando se manejan:
- muchas muestras de sensores,
- vectores de características,
- matrices de datos,
- señales discretizadas.

````python
lecturas = np.array([0.12, 0.15, 0.14, 0.20, 0.18, 0.16])

print("Arreglo:", lecturas)
print("Promedio:", np.mean(lecturas))
print("Máximo:", np.max(lecturas))
print("Mínimo:", np.min(lecturas))
print("Desviación estándar:", np.std(lecturas))
````

### Operaciones vectorizadas

NumPy permite operar sobre todos los datos al mismo tiempo.

````python
corriente = np.array([1.1, 1.3, 1.2, 1.4])
resistencia = 10

voltaje = corriente * resistencia
print("Voltajes calculados:", voltaje)
````

## 8. Arreglos bidimensionales: una base para ML

En Machine Learning es muy común representar los datos de entrada como una matriz `X`:

- cada **fila** representa una muestra,
- cada **columna** representa una característica (*feature*).

Por ejemplo, para un motor podríamos registrar:
- temperatura,
- corriente,
- vibración,
- velocidad angular.

````python
X = np.array([
    [35.2, 1.5, 0.02, 1200],
    [40.1, 1.8, 0.03, 1250],
    [55.0, 2.7, 0.08, 1300],
    [62.3, 3.1, 0.10, 1350]
])

print("Matriz de características X:")
print(X)
print("Forma de X:", X.shape)  # filas, columnas
````

### Etiquetas o salida esperada

En muchos problemas supervisados se utiliza un vector `y` para representar la clase o valor objetivo.

````python
y = np.array([0, 0, 1, 1])  # 0 = normal, 1 = posible falla

print("Etiquetas y:", y)
print("Forma de y:", y.shape)
````

## 9. Manejo de datos tabulares con pandas

`pandas` es una de las librerías más importantes para ML porque facilita:

- cargar archivos CSV,
- limpiar datos,
- seleccionar columnas,
- filtrar condiciones,
- calcular estadísticas rápidas.

````python
data = {
    "temperatura_C": [35.2, 40.1, 55.0, 62.3, 38.5],
    "corriente_A":   [1.5, 1.8, 2.7, 3.1, 1.6],
    "vibracion_g":   [0.02, 0.03, 0.08, 0.10, 0.025],
    "rpm":           [1200, 1250, 1300, 1350, 1220],
    "estado":        ["normal", "normal", "falla", "falla", "normal"]
}

df = pd.DataFrame(data)
df
````

### Exploración básica del DataFrame

````python
print("Primeras filas:")
display(df.head())

print("\nInformación general:")
print(df.info())

print("\nEstadísticos descriptivos:")
display(df.describe())
````

### Filtrado de datos

````python
muestras_falla = df[df["estado"] == "falla"]
muestras_falla
````

### Selección de características para ML

Una práctica muy común es separar:
- `X`: variables de entrada,
- `y`: variable objetivo.

````python
X_df = df[["temperatura_C", "corriente_A", "vibracion_g", "rpm"]]
y_df = df["estado"]

print("Entradas (X):")
display(X_df)

print("Salida (y):")
display(y_df)
````

## 10. Visualización de datos

Antes de entrenar un modelo de ML, conviene **ver los datos**.  
Las gráficas ayudan a detectar patrones, agrupamientos, tendencias o valores atípicos.

````python
plt.plot(df["temperatura_C"], marker="o")
plt.title("Temperatura del sistema")
plt.xlabel("Muestra")
plt.ylabel("Temperatura (°C)")
plt.grid(True)
plt.show()
````

````python
plt.scatter(df["corriente_A"], df["temperatura_C"])
plt.title("Corriente vs Temperatura")
plt.xlabel("Corriente (A)")
plt.ylabel("Temperatura (°C)")
plt.grid(True)
plt.show()
````

### Interpretación ingenieril

Si al aumentar la corriente también aumenta la temperatura, podría existir una relación entre carga eléctrica y calentamiento del sistema.  
Ese tipo de patrones es justamente el punto de partida para construir modelos predictivos.

## 11. Conceptos previos al ML: limpieza y preparación de datos

Antes de aplicar ML, normalmente se realizan tareas como:

1. verificar datos faltantes,
2. eliminar o corregir errores,
3. convertir categorías a números,
4. escalar o normalizar valores,
5. separar datos de entrenamiento y prueba.

Aquí haremos una preparación sencilla.

````python
# Revisar si hay datos faltantes
print(df.isnull().sum())
````

### Codificación simple de etiquetas

Muchos modelos necesitan datos numéricos.  
Por ello, una categoría como `"normal"` o `"falla"` puede convertirse a números.

````python
df["estado_cod"] = df["estado"].map({"normal": 0, "falla": 1})
df
````

### Conversión final a arreglos NumPy

````python
X_ml = df[["temperatura_C", "corriente_A", "vibracion_g", "rpm"]].to_numpy()
y_ml = df["estado_cod"].to_numpy()

print("X_ml:")
print(X_ml)
print("\ny_ml:")
print(y_ml)
````

## 12. Ejemplo conceptual de normalización

En ML, algunas variables pueden tener escalas muy distintas.  
Por ejemplo:
- vibración en valores pequeños,
- rpm en valores grandes.

Esto puede afectar ciertos algoritmos.  
Una forma sencilla de ilustrarlo es usar normalización Min-Max.

````python
X_min = X_ml.min(axis=0)
X_max = X_ml.max(axis=0)

X_norm = (X_ml - X_min) / (X_max - X_min)

print("Datos normalizados:")
print(np.round(X_norm, 3))
````
