Extraccion de fotogramas de video

```python
import cv2
import os

def extract_frames(video_path, output_folder, target_width=550):
    # Obtener el nombre base del archivo de video sin extensión
    video_name = os.path.splitext(os.path.basename(video_path))[0]
    
    # Crear la carpeta de salida si no existe
    os.makedirs(output_folder, exist_ok=True)
    
    # Cargar el video
    cap = cv2.VideoCapture(video_path)
    frame_count = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break  # Salir si ya no hay más fotogramas
        
        # Obtener dimensiones originales
        height, width = frame.shape[:2]

        # Calcular la nueva altura manteniendo la proporción
        aspect_ratio = height / width
        new_height = int(target_width * aspect_ratio)

        # Redimensionar la imagen
        resized_frame = cv2.resize(frame, (target_width, new_height), interpolation=cv2.INTER_AREA)

        # Definir el nombre del archivo de salida
        frame_filename = f"frame_{frame_count:05d}.jpg"
        frame_path = os.path.join(output_folder, frame_filename)

        # Guardar el fotograma como imagen
        cv2.imwrite(frame_path, resized_frame)
        frame_count += 1

    cap.release()
    print(f"Extracción completada. Se guardaron {frame_count} fotogramas en {output_folder}")

# Ejemplo de uso
video_file = "VIDEOS DATASET 2025/QRO_GLF_21_MAY_2024/GX010009/GX010009_000300_000330.mp4"  # Ruta de tu video
output_directory = "frames"  # Carpeta donde se guardarán las imágenes
extract_frames(video_file, output_directory)
```
