## 1. ¿Qué es `matplotlib`?

`matplotlib` es una biblioteca de Python que permite crear gráficas y visualizaciones.  
Es una herramienta muy útil para:

- analizar señales de sensores,
- observar tendencias en variables físicas,
- comparar experimentos,
- detectar anomalías en datos,
- presentar resultados de manera profesional.

La subbiblioteca más usada es:

```python
import matplotlib.pyplot as plt
```

`pyplot` funciona de manera similar a una pizarra de dibujo: permite crear figuras, trazar líneas, barras, puntos y mucho más.

````python
import numpy as np
import matplotlib.pyplot as plt
````

## 2. Primera gráfica: temperatura de un sensor

Supongamos que un sensor registra la temperatura de un proceso durante 10 segundos.

````python
tiempo = np.arange(0, 10, 1)
temperatura = np.array([24, 24.5, 25, 25.8, 26.2, 26.8, 27.1, 27.4, 27.8, 28.0])

plt.figure(figsize=(8,4))
plt.plot(tiempo, temperatura)
plt.title("Temperatura medida por un sensor")
plt.xlabel("Tiempo (s)")
plt.ylabel("Temperatura (°C)")
plt.grid(True)
plt.show()
````

### Explicación

- `plt.figure(figsize=(8,4))` crea una figura de tamaño 8x4.
- `plt.plot(x, y)` dibuja una gráfica de línea.
- `plt.title()` agrega un título.
- `plt.xlabel()` y `plt.ylabel()` etiquetan los ejes.
- `plt.grid(True)` agrega una cuadrícula.
- `plt.show()` muestra la figura.

Esta es una de las gráficas más utilizadas cuando se desea ver cómo cambia una variable con el tiempo.

## 3. Personalización de una gráfica

Podemos modificar la apariencia para hacer la visualización más clara.

````python
plt.figure(figsize=(9,4))
plt.plot(tiempo, temperatura, marker='o', linestyle='--', linewidth=2, label='Sensor T1')
plt.title("Temperatura del sistema")
plt.xlabel("Tiempo (s)")
plt.ylabel("Temperatura (°C)")
plt.legend()
plt.grid(True)
plt.show()
````

### Elementos importantes

- `marker='o'` coloca puntos sobre cada dato.
- `linestyle='--'` cambia el estilo de línea.
- `linewidth=2` aumenta el grosor.
- `label='Sensor T1'` asigna nombre a la serie.
- `plt.legend()` muestra la leyenda.

En reportes técnicos, esta personalización ayuda a distinguir señales y mejora la interpretación.

## 4. Comparación de dos sensores en una misma gráfica

En mecatrónica es común comparar dos o más señales al mismo tiempo.

````python
temperatura_sensor_2 = np.array([23.8, 24.2, 24.7, 25.1, 25.7, 26.1, 26.6, 27.0, 27.3, 27.6])

plt.figure(figsize=(9,4))
plt.plot(tiempo, temperatura, marker='o', label='Sensor 1')
plt.plot(tiempo, temperatura_sensor_2, marker='s', label='Sensor 2')
plt.title("Comparación de temperatura entre dos sensores")
plt.xlabel("Tiempo (s)")
plt.ylabel("Temperatura (°C)")
plt.legend()
plt.grid(True)
plt.show()
````

### Interpretación

Con esta gráfica podemos observar:

- cuál sensor reporta valores mayores,
- si ambos siguen la misma tendencia,
- si existe desfase o diferencia de calibración.

Este tipo de gráfica es muy útil para comparar sensores redundantes o validar instrumentos.

## 5. Gráfica de dispersión (`scatter`)

La gráfica de dispersión se usa para observar la relación entre dos variables.

Ejemplo: relación entre **voltaje** y **velocidad** en un motor de corriente directa.

````python
voltaje = np.array([2, 4, 6, 8, 10, 12])
velocidad = np.array([120, 250, 360, 470, 590, 710])

plt.figure(figsize=(7,4))
plt.scatter(voltaje, velocidad, s=80)
plt.title("Relación entre voltaje y velocidad del motor")
plt.xlabel("Voltaje (V)")
plt.ylabel("Velocidad (rpm)")
plt.grid(True)
plt.show()
````

### ¿Para qué sirve?

Sirve para detectar si existe una relación entre dos variables.  
Por ejemplo, en este caso se observa que al aumentar el voltaje, la velocidad también aumenta.

Esta visualización es muy utilizada en:

- caracterización de actuadores,
- análisis experimental,
- exploración de datos previa a modelos de ML.

## 6. Gráfica de barras

La gráfica de barras es útil para comparar cantidades discretas.

Ejemplo: error promedio de tres controladores.

````python
controladores = ['P', 'PI', 'PID']
error_promedio = [8.5, 4.2, 2.1]

plt.figure(figsize=(7,4))
plt.bar(controladores, error_promedio)
plt.title("Comparación de error promedio por controlador")
plt.xlabel("Controlador")
plt.ylabel("Error promedio")
plt.grid(axis='y')
plt.show()
````

### Uso típico en ingeniería

Las barras se usan para comparar:

- consumo de energía entre sistemas,
- error de diferentes algoritmos,
- tiempo de respuesta de varios métodos,
- cantidad de fallas por máquina o estación.

## 7. Histograma

El histograma sirve para observar cómo se distribuyen los valores de una variable.

Ejemplo: mediciones de distancia de un sensor ultrasónico.

````python
np.random.seed(10)
distancias = np.random.normal(loc=50, scale=4, size=100)

plt.figure(figsize=(8,4))
plt.hist(distancias, bins=12, edgecolor='black')
plt.title("Histograma de mediciones de distancia")
plt.xlabel("Distancia (cm)")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
````

### ¿Qué nos permite ver?

- valores frecuentes,
- dispersión de datos,
- posibles sesgos,
- presencia de ruido o mediciones atípicas.

Este tipo de gráfica es muy valiosa cuando se analiza el comportamiento de sensores.

## 8. Subgráficas

A veces es conveniente mostrar varias visualizaciones en una sola figura.

````python
corriente = np.array([1.2, 1.4, 1.6, 1.8, 2.0, 2.1, 2.3, 2.5, 2.6, 2.8])

fig, axs = plt.subplots(2, 1, figsize=(8,6))

axs[0].plot(tiempo, temperatura, marker='o')
axs[0].set_title("Temperatura del sistema")
axs[0].set_xlabel("Tiempo (s)")
axs[0].set_ylabel("Temperatura (°C)")
axs[0].grid(True)

axs[1].plot(tiempo, corriente, marker='s')
axs[1].set_title("Corriente del actuador")
axs[1].set_xlabel("Tiempo (s)")
axs[1].set_ylabel("Corriente (A)")
axs[1].grid(True)

plt.tight_layout()
plt.show()
````

### Ventaja de las subgráficas

Permiten comparar visualmente distintas variables del mismo sistema, por ejemplo:

- temperatura y corriente,
- posición y velocidad,
- referencia y salida,
- señal de entrada y señal filtrada.

## 9. Ejemplo aplicado a una señal senoidal

En mecatrónica, muchas señales pueden modelarse o analizarse como ondas.  
Por ejemplo, vibraciones, señales periódicas o referencias.

````python
t = np.linspace(0, 2*np.pi, 200)
senal = np.sin(t)

plt.figure(figsize=(8,4))
plt.plot(t, senal)
plt.title("Señal senoidal")
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()
````

### Aplicaciones

Esta gráfica puede representar:

- una vibración mecánica ideal,
- una señal de prueba,
- una referencia de movimiento,
- el comportamiento oscilatorio de un sistema.

## 10. Comparación entre señal ideal y señal con ruido

En sistemas reales, los datos de sensores suelen contener ruido.

````python
np.random.seed(5)
senal_ruidosa = senal + np.random.normal(0, 0.15, len(senal))

plt.figure(figsize=(9,4))
plt.plot(t, senal, label='Señal ideal')
plt.plot(t, senal_ruidosa, label='Señal con ruido')
plt.title("Comparación entre señal ideal y señal ruidosa")
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.legend()
plt.grid(True)
plt.show()
````

### Relación con ML

Antes de entrenar un modelo de Machine Learning, es importante **visualizar los datos** para:

- detectar ruido,
- identificar valores fuera de rango,
- observar tendencias,
- decidir si se requiere limpieza o filtrado.

## 11. Guardar una figura

Las figuras pueden guardarse para reportes o presentaciones.

````python
plt.figure(figsize=(8,4))
plt.plot(tiempo, temperatura, marker='o')
plt.title("Temperatura del sistema")
plt.xlabel("Tiempo (s)")
plt.ylabel("Temperatura (°C)")
plt.grid(True)

# Guardar la imagen
plt.savefig("temperatura_sistema.png", dpi=300, bbox_inches='tight')
plt.show()
````

### Parámetros importantes

- `dpi=300` mejora la resolución.
- `bbox_inches='tight'` reduce márgenes innecesarios.

Esto es útil para insertar figuras en informes, artículos o presentaciones.
