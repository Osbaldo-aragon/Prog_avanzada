## 1. ¿Qué es la regresión lineal?

La regresión lineal es un método que busca encontrar una **línea recta** que describa la relación entre dos variables.

La ecuación general es:

\[
y = mx + b
\]

Donde:

- **x**: variable de entrada o independiente,
- **y**: variable de salida o dependiente,
- **m**: pendiente de la recta,
- **b**: intercepto con el eje y.

### Interpretación
- La **pendiente** indica cuánto cambia `y` cuando `x` aumenta una unidad.
- El **intercepto** indica el valor de `y` cuando `x = 0`.

Si la nube de puntos sigue una tendencia recta, la regresión lineal puede describir esa relación.


```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Configuración simple para mostrar tablas completas
pd.set_option("display.precision", 3)
```


## 2. Ejemplo sencillo: Voltaje aplicado vs velocidad de un motor DC

Supongamos que se realizaron varias mediciones en un motor de corriente directa y se obtuvieron los siguientes datos:

- **Entrada (x):** voltaje aplicado en volts
- **Salida (y):** velocidad en RPM

Estos datos son ficticios pero realistas para fines didácticos.


```python

voltaje = np.array([2, 4, 6, 8, 10, 12], dtype=float)
rpm = np.array([180, 390, 610, 790, 980, 1190], dtype=float)

df = pd.DataFrame({
    "Voltaje_V": voltaje,
    "Velocidad_RPM": rpm
})

df
```


## 3. Visualización de los datos

Antes de aplicar cualquier modelo, siempre es recomendable observar los datos.


```python

plt.figure(figsize=(8,5))
plt.scatter(voltaje, rpm)
plt.title("Voltaje vs Velocidad del motor")
plt.xlabel("Voltaje (V)")
plt.ylabel("Velocidad (RPM)")
plt.grid(True)
plt.show()
```


La gráfica muestra una tendencia creciente. A medida que aumenta el voltaje, también aumenta la velocidad del motor.  
Esto sugiere que una recta puede ser una buena primera aproximación.



## 4. Ajuste de una recta de forma simple con `numpy`

Podemos estimar una recta usando `numpy.polyfit()`.


```python

m, b = np.polyfit(voltaje, rpm, 1)

print(f"Pendiente m = {m:.3f}")
print(f"Intercepto b = {b:.3f}")
```


Con esos valores, el modelo lineal queda de la forma:

\[
\text{RPM} = m \cdot \text{Voltaje} + b
\]

Ahora usamos esa ecuación para calcular la recta ajustada.


```python

rpm_estimadas = m * voltaje + b

tabla_modelo = pd.DataFrame({
    "Voltaje_V": voltaje,
    "RPM_real": rpm,
    "RPM_estimada": rpm_estimadas
})

tabla_modelo
```

```python

plt.figure(figsize=(8,5))
plt.scatter(voltaje, rpm, label="Datos reales")
plt.plot(voltaje, rpm_estimadas, label="Recta ajustada")
plt.title("Regresión lineal simple")
plt.xlabel("Voltaje (V)")
plt.ylabel("Velocidad (RPM)")
plt.grid(True)
plt.legend()
plt.show()
```


## 5. Predicción con el modelo

Una vez obtenida la ecuación, podemos estimar la velocidad para un voltaje no medido directamente.

Por ejemplo: ¿qué velocidad se espera a **9 V**?


```python

voltaje_nuevo = 9
rpm_predicha = m * voltaje_nuevo + b

print(f"Para {voltaje_nuevo} V, la velocidad estimada es {rpm_predicha:.2f} RPM")
```


## 6. Error básico del modelo

Para saber qué tan bien se ajusta la recta a los datos, podemos calcular el error entre los valores reales y los estimados.


```python

error = rpm - rpm_estimadas
error_abs = np.abs(error)

tabla_error = pd.DataFrame({
    "Voltaje_V": voltaje,
    "RPM_real": rpm,
    "RPM_estimada": rpm_estimadas,
    "Error": error,
    "Error_absoluto": error_abs
})

tabla_error
```


Si los errores son pequeños y la recta sigue bien la tendencia de los datos, entonces el modelo puede considerarse útil para una primera aproximación.



## 7. Uso con `scikit-learn`

En Machine Learning, una forma muy común de trabajar la regresión lineal es con `LinearRegression` de `scikit-learn`.


```python

from sklearn.linear_model import LinearRegression

X = voltaje.reshape(-1, 1)   # variable de entrada en formato columna
y = rpm                      # variable de salida

modelo = LinearRegression()
modelo.fit(X, y)

print("Pendiente:", modelo.coef_[0])
print("Intercepto:", modelo.intercept_)
```

```python

rpm_sklearn = modelo.predict(X)

plt.figure(figsize=(8,5))
plt.scatter(voltaje, rpm, label="Datos reales")
plt.plot(voltaje, rpm_sklearn, label="Modelo sklearn")
plt.title("Regresión lineal con scikit-learn")
plt.xlabel("Voltaje (V)")
plt.ylabel("Velocidad (RPM)")
plt.grid(True)
plt.legend()
plt.show()
```


## 8. Coeficiente de determinación \(R^2\)

Una medida muy común para evaluar el ajuste es el coeficiente \(R^2\).

- Si \(R^2\) está cerca de **1**, el modelo explica muy bien la variación de los datos.
- Si \(R^2\) está cerca de **0**, el modelo explica muy poco.


```python

r2 = modelo.score(X, y)
print(f"R^2 = {r2:.4f}")
```


## 9. Interpretación en lenguaje sencillo

Si la pendiente es positiva:

- al aumentar el voltaje, aumenta la velocidad.

Si la pendiente fuera negativa:

- al aumentar la entrada, la salida disminuiría.

En este ejemplo, la pendiente positiva tiene sentido físico, porque un motor DC generalmente incrementa su velocidad cuando aumenta el voltaje de alimentación, al menos dentro de cierto rango de operación.



## 10. Otro ejemplo breve: temperatura y ventilación

Imaginemos un sistema donde se registra la temperatura de una cámara y la velocidad del ventilador.


```python

temperatura = np.array([25, 28, 30, 33, 35, 38, 40], dtype=float)
ventilador = np.array([20, 28, 35, 48, 57, 70, 78], dtype=float)  # porcentaje PWM

m2, b2 = np.polyfit(temperatura, ventilador, 1)
vent_est = m2 * temperatura + b2

plt.figure(figsize=(8,5))
plt.scatter(temperatura, ventilador, label="Datos")
plt.plot(temperatura, vent_est, label="Ajuste lineal")
plt.title("Temperatura vs velocidad del ventilador")
plt.xlabel("Temperatura (°C)")
plt.ylabel("PWM del ventilador (%)")
plt.grid(True)
plt.legend()
plt.show()

print(f"Ecuación aproximada: y = {m2:.2f}x + {b2:.2f}")
```


Este tipo de modelo puede servir como base para:

- estimar comportamientos,
- diseñar reglas de control,
- detectar tendencias,
- construir modelos más avanzados posteriormente.



## 11. Ventajas de la regresión lineal

- Es sencilla de entender e implementar.
- Es útil como primer modelo de análisis.
- Permite hacer predicciones rápidas.
- Ayuda a identificar tendencias entre variables.

## 12. Limitaciones

- No siempre los datos siguen una relación lineal.
- Es sensible a valores atípicos.
- No captura comportamientos no lineales complejos.
- En sistemas reales, una variable puede depender de varias entradas y no solo de una.



## 13. Conclusión

La regresión lineal es una herramienta básica pero muy útil para empezar en Machine Learning y análisis de datos.

Para alumnos de Ingeniería Mecatrónica, este método permite modelar relaciones simples entre variables de sensores, actuadores o procesos. También sirve como base para comprender métodos más avanzados.
