## Ejercicio 1. Registro y análisis básico de temperatura

### Contexto
En un sistema de monitoreo, se registran temperaturas de un actuador eléctrico durante varios instantes.

### Datos de ejemplo


```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

temperatura = [28, 29, 30, 31, 31, 32, 33, 34, 35, 36]
tiempo = list(range(1, 11))

df_temp = pd.DataFrame({
    "Tiempo_s": tiempo,
    "Temperatura_C": temperatura
})

df_temp
```


### Actividades
1. Mostrar la tabla de datos.
2. Calcular:
   - media,
   - valor máximo,
   - valor mínimo.
3. Graficar temperatura contra tiempo.
4. Interpretar si la tendencia parece estable o creciente.
5. Redactar una conclusión breve.



## Ejercicio 2. Análisis de vibración de un motor

### Contexto


### Actividades

## Ejercicio 3. Uso de `numpy` con corriente y voltaje

### Contexto
Se tienen arreglos con mediciones de voltaje y corriente en un actuador.

### Datos de ejemplo


```python

voltaje = np.array([12, 12, 11.9, 12.1, 12.0, 11.8, 12.2])
corriente = np.array([1.2, 1.3, 1.25, 1.35, 1.28, 1.4, 1.38])

print("Voltaje promedio:", voltaje.mean())
print("Corriente promedio:", corriente.mean())
```


### Actividades

2. Calcular el valor máximo y mínimo de ambas variables.
3. Graficar ambas señales.
4. Explicar por qué `numpy` es útil para trabajar con mediciones experimentales.
