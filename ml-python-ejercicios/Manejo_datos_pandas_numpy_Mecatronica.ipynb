## 1. Importación de bibliotecas
En esta sección importamos las herramientas principales que usaremos.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Configuración opcional para mostrar más columnas
pd.set_option("display.max_columns", None)
pd.set_option("display.precision", 3)
```

## 2. Introducción a NumPy

NumPy trabaja con arreglos numéricos llamados `ndarray`.  
Estos arreglos son más eficientes que las listas tradicionales de Python para cálculos matemáticos.

### 2.1 Crear arreglos

```python
# Arreglo unidimensional: por ejemplo, lecturas de temperatura en °C
temperaturas = np.array([24.5, 25.1, 25.8, 26.0, 25.7, 25.3])
print("Temperaturas:", temperaturas)
print("Tipo:", type(temperaturas))
print("Dimensiones:", temperaturas.ndim)
print("Forma:", temperaturas.shape)
```

### 2.2 Operaciones matemáticas básicas

Una ventaja de NumPy es que puede operar directamente con todos los elementos del arreglo.

```python
voltajes = np.array([4.8, 4.9, 5.0, 5.1, 5.0])

print("Voltajes originales:", voltajes)
print("Voltajes + 1:", voltajes + 1)
print("Voltajes * 2:", voltajes * 2)
print("Voltajes al cuadrado:", voltajes**2)
```

### 2.3 Estadísticos básicos con NumPy

```python
corriente_motor = np.array([1.2, 1.4, 1.3, 1.5, 1.7, 1.6, 1.4])

print("Promedio:", np.mean(corriente_motor))
print("Máximo:", np.max(corriente_motor))
print("Mínimo:", np.min(corriente_motor))
print("Desviación estándar:", np.std(corriente_motor))
```

### 2.4 Crear secuencias útiles

Estas funciones son muy comunes cuando trabajamos con muestras o ejes de tiempo.

```python
# 10 muestras igualmente espaciadas entre 0 y 5 segundos
tiempo = np.linspace(0, 5, 10)

# Secuencia de enteros
muestras = np.arange(0, 10, 1)

print("Tiempo:", tiempo)
print("Muestras:", muestras)
```

### 2.5 Ejemplo: señal simple de vibración

Simularemos una señal de vibración senoidal, útil para entender cómo representar datos medidos en el tiempo.

```python
t = np.linspace(0, 2, 200)   # 2 segundos, 200 puntos
frecuencia = 5               # 5 Hz
amplitud = 2

vibracion = amplitud * np.sin(2 * np.pi * frecuencia * t)

plt.figure(figsize=(10, 4))
plt.plot(t, vibracion)
plt.title("Señal de vibración simulada")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()
```

## 3. Introducción a pandas

Mientras que NumPy es excelente para cálculo numérico, **pandas** es ideal para trabajar con datos tabulares.

La estructura principal en pandas es el **DataFrame**, que se parece a una hoja de cálculo.

### 3.1 Crear un DataFrame manualmente

```python
datos_sensores = {
    "tiempo_s": [0, 1, 2, 3, 4, 5],
    "temperatura_C": [28.1, 28.3, 28.8, 29.2, 29.5, 29.7],
    "corriente_A": [1.10, 1.12, 1.15, 1.20, 1.18, 1.22],
    "rpm": [1450, 1452, 1458, 1460, 1459, 1461]
}

df = pd.DataFrame(datos_sensores)
df
```

### 3.2 Inspección básica del DataFrame

```python
print("Primeras filas:")
display(df.head())

print("Últimas filas:")
display(df.tail())

print("Información general:")
df.info()

print("\nResumen estadístico:")
display(df.describe())
```

### 3.3 Selección de columnas

```python
# Una columna
display(df["temperatura_C"])

# Varias columnas
display(df[["temperatura_C", "corriente_A"]])
```

### 3.4 Filtrado de datos
Podemos seleccionar solo los registros que cumplan una condición.

```python
# Filtrar registros donde la temperatura sea mayor a 29 °C
temperatura_alta = df[df["temperatura_C"] > 29]
temperatura_alta
```

### 3.5 Crear nuevas columnas
Esto es muy útil para ingeniería, por ejemplo para calcular potencia aproximada o convertir unidades.

```python
# Potencia eléctrica aproximada en watts asumiendo 24 V
df["potencia_W"] = 24 * df["corriente_A"]

# Conversión simple de temperatura a Kelvin
df["temperatura_K"] = df["temperatura_C"] + 273.15

df
```

## 4. Relación entre NumPy y pandas

pandas está construido sobre NumPy, por eso ambas bibliotecas se complementan muy bien.

Podemos extraer columnas de pandas como arreglos NumPy.

```python
temperaturas_np = df["temperatura_C"].to_numpy()
print(type(temperaturas_np))
print(temperaturas_np)
```

## 5. Ejemplo aplicado: análisis de un proceso mecatrónico

Vamos a simular un conjunto de datos más realista de una banda transportadora con:

- tiempo,
- temperatura del motor,
- corriente,
- velocidad,
- estado del sistema.

```python
np.random.seed(42)

n = 50
tiempo = np.arange(n)

temperatura = 35 + np.random.normal(0, 1.5, n).cumsum() * 0.1
corriente = 2.5 + np.random.normal(0, 0.15, n)
velocidad = 1200 + np.random.normal(0, 20, n)
estado = np.where(corriente > 2.7, "alerta", "normal")

df_proceso = pd.DataFrame({
    "tiempo_s": tiempo,
    "temperatura_C": temperatura,
    "corriente_A": corriente,
    "velocidad_rpm": velocidad,
    "estado": estado
})

df_proceso.head()
```

### 5.1 Consultar dimensiones y columnas

```python
print("Forma:", df_proceso.shape)
print("Columnas:", df_proceso.columns.tolist())
```

### 5.2 Filtrar condiciones de alerta

```python
df_alerta = df_proceso[df_proceso["estado"] == "alerta"]
df_alerta.head()
```

### 5.3 Promedios y agrupación
En muchos casos interesa resumir valores por categoría.

```python
resumen_estado = df_proceso.groupby("estado")[["temperatura_C", "corriente_A", "velocidad_rpm"]].mean()
resumen_estado
```

## 6. Limpieza básica de datos

En la práctica, los datos reales pueden tener:

- valores faltantes,
- mediciones erróneas,
- formatos incorrectos,
- ruido o duplicados.

Veamos algunos casos básicos.

### 6.1 Crear un ejemplo con datos faltantes

```python
df_limpieza = df_proceso.copy()

# Introducir algunos valores faltantes de ejemplo
df_limpieza.loc[5, "temperatura_C"] = np.nan
df_limpieza.loc[12, "corriente_A"] = np.nan
df_limpieza.loc[20, "velocidad_rpm"] = np.nan

df_limpieza.head(15)
```

### 6.2 Detectar valores faltantes

```python
df_limpieza.isna().sum()
```

### 6.3 Eliminar filas con valores faltantes

```python
df_sin_nan = df_limpieza.dropna()
print("Forma original:", df_limpieza.shape)
print("Forma sin NaN:", df_sin_nan.shape)
```

### 6.4 Rellenar valores faltantes
A veces no conviene eliminar filas. Una alternativa es rellenar usando media, mediana o interpolación.

```python
df_relleno = df_limpieza.copy()

df_relleno["temperatura_C"] = df_relleno["temperatura_C"].fillna(df_relleno["temperatura_C"].mean())
df_relleno["corriente_A"] = df_relleno["corriente_A"].fillna(df_relleno["corriente_A"].mean())
df_relleno["velocidad_rpm"] = df_relleno["velocidad_rpm"].interpolate()

df_relleno.head(15)
```

## 7. Ordenamiento y consulta de datos

```python
# Ordenar por corriente de mayor a menor
df_ordenado = df_proceso.sort_values(by="corriente_A", ascending=False)
df_ordenado.head()
```

```python
# Consultar solo ciertas columnas y filas con temperatura superior a 35.5 °C
consulta = df_proceso.loc[df_proceso["temperatura_C"] > 35.5, ["tiempo_s", "temperatura_C", "corriente_A"]]
consulta.head()
```

## 8. Operaciones numéricas útiles con NumPy

En Mecatrónica es común trabajar con arreglos para cálculos rápidos.

### 8.1 Normalización de datos

```python
senal = np.array([10, 15, 20, 18, 25, 30, 28], dtype=float)

senal_norm = (senal - np.min(senal)) / (np.max(senal) - np.min(senal))

print("Señal original:", senal)
print("Señal normalizada:", senal_norm)
```

### 8.2 Cálculo de error respecto a una referencia

```python
referencia = 1500
rpm_medidas = np.array([1480, 1495, 1502, 1510, 1498])

error = referencia - rpm_medidas

print("RPM medidas:", rpm_medidas)
print("Error:", error)
print("Error promedio:", np.mean(error))
```

## 12. Preparación de datos para análisis posterior o ML

Antes de aplicar Machine Learning, normalmente se requiere:

1. seleccionar variables,
2. limpiar datos,
3. convertir categorías a números si es necesario,
4. revisar escalas,
5. separar entradas y salidas.

Ejemplo sencillo:

```python
# Copia de trabajo
df_ml = df_proceso.copy()

# Convertir estado a variable numérica
df_ml["estado_binario"] = df_ml["estado"].map({"normal": 0, "alerta": 1})

# Seleccionar variables de entrada X y salida y
X = df_ml[["temperatura_C", "corriente_A", "velocidad_rpm"]]
y = df_ml["estado_binario"]

print("Variables de entrada X:")
display(X.head())

print("Variable objetivo y:")
display(y.head())
```
